```python
import pandas as pd
import numpy as np
```

```python
df_= pd.read_csv('sop.csv')
df_.head(10)
```

```python
df=df_.drop(['Indicator Name'], axis=1)
```

```python
for col in df:
    
         df[col].fillna(df[col].mean(),inplace=True)   
```

```python
desc=df.describe().transpose()
```

```python
desc=desc.drop(['count'],axis=1)
```

```python
desc.round(decimals=3)
desc.head(25)
```

```python
df.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(9, 7))
sns.heatmap(df.corr(), cmap='RdBu')
```

```python
model=df
```

```python
x=model.drop(['GDP per capita growth (annual %)'], axis=1)
X=model.drop(['GDP per capita growth (annual %)'], axis=1).values
y=model['GDP per capita growth (annual %)'].values
```

```python
from sklearn import preprocessing
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
X_scale = min_max_scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
SEED=1
X_train,X_test,y_train,y_test=train_test_split(X_scale,y,test_size=0.2,random_state=SEED)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error as MSE
```

```python
reg=LinearRegression()
reg.fit(X_train,y_train)
y_pred=reg.predict(X_test)
accuracy_reg=reg.score(X_test,y_test)
print("Linear Regression accuracy: {}".format(accuracy_reg))
mse_reg= MSE(y_test, y_pred)
rmse_reg= mse_reg**(1/2)
print("Linear Regression rmse: {}".format(rmse_reg))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=7, random_state=SEED, n_jobs=-1)
rf.fit(X_train, y_train)
y_pred=rf.predict(X_test)
accuracy_rf=rf.score(X_test,y_test)
print("Random Forest accuracy: {}".format(accuracy_rf))
mse_rf= MSE(y_test, y_pred)
rmse= mse_rf**(1/2)
print("Random Forest rmse: {}".format(rmse))
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import BaggingRegressor
```

```python
dt= DecisionTreeRegressor(max_depth=4, min_samples_leaf=0.16, random_state=1)
bc= BaggingRegressor(base_estimator=dt, n_estimators=10, n_jobs=-1)
```

```python
dt.fit(X_train, y_train)
y_pred=dt.predict(X_test)
accuracy_dt=dt.score(X_test,y_test)
print("Decision Tree accuracy: {}".format(accuracy_dt))
mse_dt=MSE(y_test, y_pred)
rmse_dt= mse_dt**(1/2)
print("Decision Tree rmse: {}".format(rmse_dt))
```

```python
bc.fit(X_train, y_train)
y_pred=bc.predict(X_test)
accuracy_bc=bc.score(X_test,y_test)
print("Bagging accuracy: {}".format(accuracy_bc))
mse_bc= MSE(y_test, y_pred)
rmse= mse_bc**(1/2)
print("Bagging rmse: {}".format(rmse))
```
